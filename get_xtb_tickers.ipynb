```python
def xtb_parser(link, ticketwriter):
    
    response = requests.get(link)
    json_parsed = json.loads(response.text)
    instruments_to_parse = json_parsed['instrumentsCollectionLimited']['cashstocks']['US']['instruments']
    
    for key in instruments_to_parse.keys():

        instrument_details = instruments_to_parse[key]
        symbol = instrument_details['symbol']
        full_name = instrument_details['description']

        tickerwriter.writerow([symbol, full_name])
    
```
